### Handshake problem
In a room of n people, how many different handshakes are possible? 

Analytically: 

This is a combinations problem.

<img src="images/combo.png" style="width: 200px;"/>

C(n,2) = n!/((n-2)! 2!) = n(n-1)/2

Python does this easily!


```python
from itertools import combinations

handshakes = list(combinations(['Aaron','John','Betsy', 'Steve'],2))
for handshake in handshakes:
    print(handshake)
print(len(handshakes))
```

To be more general about it, we could use a general n people:

```python
from itertools import combinations

n=4
handshakes = list(combinations(range(n),2))
for handshake in handshakes:
    print(handshake)
print(len(handshakes))
```

Of course, we can also obtain this using the factorial formula:

```python
import math
math.factorial(4)/(math.factorial(2)*math.factorial(4-2))
```

How many unique 7 digit license plates can be made out of A,B,C,D,1,2,3,4?

This is a permutations problem.

<img src="images/perm.png" style="width: 200px;"/>

P(8,7) = 8!/(8-7)! = 8!/1! = 40320.

Python can do this easily, too.

```python
from itertools import permutations

licenses = list(permutations(['A','B','C','D','1','2','3','4'],7))
for license in licenses:
    print(license)
print(len(licenses))
```

Of course, we can also obtain this using the factorial formula:

```python
import math
math.factorial(8)/math.factorial(8-7)
```

```python
import random
students = ['Jack', 'Abby', 'Mary', 'Peter', "Stella"]
print(random.sample(students,2))
```

Suppose you want to simulate rolling a die and add each die roll to a list. Yes, you can create a list of numbers 1,2,3,4,5, and 6, but if you do rand(sample, 3), then you'll never get the same number twice, which doesn't make sense, since the rolls of the dice are independent of each other. To randomly generate ten rolls and add them to a list, where some numbers may be the same, type:

```python
import random
rolls = []
for i in range(10):
    rolls.append(random.randint(1,6))
print(rolls)
```

What if it wasn't a fair die? What if the probability of getting each number was given by:

|  Die Face |Probability   |
|---   |---   |
|1 |0.05  | 
|2 | 0.05 |
|3 |0.50  |
|4 | 0.10 |
|5 |0.20  |
|6 | 0.10 |

Then we would select numbers between 0 and 99 based on the above probabilities. For example, since there is a 5% chance that we will roll a 1, we'll assign the numbers 0,1,2,3, and 4 to the die roll 1. Here are fifty rolls:

```python
import random
rolls = []
for i in range(50):
    roll = random.randint(0,99)
    if 0 <= roll < 5:
        rolls.append(1)
    if 5 <= roll < 10:
        rolls.append(2)
    if 10 <= roll < 60:
        rolls.append(3)
    if 60 <= roll < 70:
        rolls.append(4)
    if 70 <= roll < 90:
        rolls.append(5)
    if 90 <= roll < 100:
        rolls.append(6)
print(rolls)
```

What was our average die roll?

```python
print(sum(rolls)/len(rolls))
```

### Expected Value

Let's go back to a fair die:

|  Die Face |Probability   |
|---   |---   |
|1 |1/6  | 
|2 | 1/6 |
|3 |1/6  |
|4 | 1/6 |
|5 |1/6  |
|6 | 1/6 |


We can calculate the expected value of the die face by adding the products of the face value times the probabilities of getting that face value: (1)(1)(1/6)+(2)(1/6)+(3)(1/6)+(4)(1/6)+(5)(1/6)+(6)(1/6) = 3.5.

Consider the following scenario. You bet a dollar to play a game. If you roll a die and get a six, then you win 5 dollars. If you roll any other number, you lose your dollar. What are the theoretical probabilities?

|  Earnings |Probability   |
|---   |---   |
|-1 |5/6  | 
|5 | 1/6 |

Now, the expected value of your earnings would be (-1)(5/6)+(5)(1/6) = 0 dollars.

We can simulate this game numerically. We'll randomly roll a die (i.e., generate a random number between 1 and 6). If we roll a 6, we'll give ourselves five dollars; otherwise, we'll lose a dollar. We'll roll the die 10000 times and see that we get an experimental average earning of close to zero:

```python
import random

earnings = 0
for i in range(10000):
    die = random.randint(1,6)
    if die == 6:
        earnings = earnings + 5
    else:
        earnings = earnings - 1
print(earnings/10000)
```

What if you wanted to keep track of your number of rolls versus your average earnings? You could make two lists and then plot them.

```python
import random
import matplotlib.pyplot as plt
%matplotlib inline

earnings = 0
avg_earnings = [0]
number_of_sims = [0]

for i in range(10000):
    
    die = random.randint(1,6)
    
    if die == 6:
        earnings = earnings + 5
    else:
        earnings = earnings - 1

    number_of_sims.append(number_of_sims[-1]+1)
    avg_earnings.append(earnings/number_of_sims[-1])
    
plt.plot(number_of_sims, avg_earnings,'.')
```
