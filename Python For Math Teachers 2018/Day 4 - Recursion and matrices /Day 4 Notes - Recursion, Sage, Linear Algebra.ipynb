To make plots side by side:

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['figure.figsize'] = 10,10

x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2,2,1)
plt.plot(x1, y1, 'ko-')
plt.title('Damped oscillation')
plt.ylabel('y (cm)')

plt.subplot(2,1,2)
plt.plot(x2, y2, 'r.-')
plt.title('Undamped oscillation')
plt.xlabel('time (s)')
plt.ylabel('y (cm)')
```

### Linear Algebra
You can perform many linear algebra operations using the linalg package inside scipy. 
You can create matrices:

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import linalg

A=np.matrix([[1,2,2], [4,4,2],[4,6,4]])
B=np.matrix([[3],[6],[10]])

#or populate cells indivually
#A = np.zeros((3,2))
#A[0][1] = 2
#print(A)
```

Multiply, find the determinant, and find the inverse of matrices:

```python
print('A*B')
print(A*B)
print('det(A)')
print(linalg.det(A))
print('inv(A)')
print(linalg.inv(A))
```

Solve the matrix equation Ax = B two different ways:

```python
print('using the inverse to solve:')
print(linalg.inv(A).dot(B))
print('using solve:')
linalg.solve(A,B)
```

And find eigenvalues and eigenvectors:

```python
eigenvalues, eigenvectors = linalg.eig(A)
print(eigenvalues)
```
