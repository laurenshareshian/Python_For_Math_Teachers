### Handshake problem
In a room of n people, how many different handshakes are possible? 

Analytically: 

This is a combinations problem.

$C(n,2) = \frac{n!}{(n-2)! 2!} = \frac{n(n-1)}{2}$

Python does this easily!


```python
from itertools import combinations

handshakes = list(combinations(['Aaron','John','Betsy','Steve'],2))
for handshake in handshakes:
    print(handshake)
print(f"Number of handshakes: {len(handshakes)}")
```

To be more general about it, we could use a general n people:

```python
from itertools import combinations

n=4
handshakes = list(combinations(range(n),2))
for handshake in handshakes:
    print(handshake)
print(len(handshakes))
print(f"Number of handshakes: {len(handshakes)}")
```

Of course, we can also obtain this using the factorial formula:

$C(n,2) = \frac{n!}{(n-2)! 2!} = \frac{n(n-1)}{2}$

```python
import math
math.factorial(4)/(math.factorial(2)*math.factorial(4-2))
```

How many unique 7 digit license plates can be made out of A,B,C,D,1,2,3,4?

This is a permutations problem.

$P(8,7) = \frac{8!}{(8-7)!} = \frac{8!}{1!} = 40320.$

Python can do this easily, too.

```python
from itertools import permutations

licenses = list(permutations(['A','B','C','D','1','2','3','4'],7))
for license in licenses:
    print(license)
print(f"Number of licenses: {len(licenses)}")
```

Of course, we can also obtain this using the factorial formula:

$P(8,7) = \frac{8!}{(8-7)!} = \frac{8!}{1!} = 40320$

```python
import math
math.factorial(8)/math.factorial(8-7)
```

### Expected Value

Suppose we have fair dice:

|  Die Face |Probability   |
|---   |---   |
|1 |1/6  | 
|2 | 1/6 |
|3 |1/6  |
|4 | 1/6 |
|5 |1/6  |
|6 | 1/6 |


We can calculate the expected value of the die face by adding the products of the face value times the probabilities of getting that face value: 

$(1)(1)(1/6)+(2)(1/6)+(3)(1/6)+(4)(1/6)+(5)(1/6)+(6)(1/6) = 3.5.$

Consider the following scenario. You bet a dollar to play a game. If you roll a die and get a six, then you win 5 dollars and I get my dollar back. If you roll any other number, you lose your dollar. What are the theoretical probabilities?

|  Earnings |Probability   |
|---   |---   |
|-1 |5/6  | 
|5 | 1/6 |

Now, the expected value of your earnings would be $(-1)(5/6)+(5)(1/6) = 0$ dollars.

We can simulate this game numerically. We'll randomly roll a die (i.e., generate a random number between 1 and 6). If we roll a 6, we'll give ourselves five dollars; otherwise, we'll lose a dollar. We'll roll the die 10000 times and see that we get an experimental average earning of close to zero:

```python
earnings = 0
for i in range(10000):
    die = random.randint(1,6)
    if die == 6:
        earnings = earnings + 5
    else:
        earnings = earnings - 1
print(earnings/10000)
```

What if you wanted to keep track of your number of rolls versus your average earnings? You could make two lists and then plot them.

```python
earnings = 0
avg_earnings = [0]
number_of_sims = [0]

for i in range(10000):
    
    die = random.randint(1,6)
    
    if die == 6:
        earnings = earnings + 5
    else:
        earnings = earnings - 1

    number_of_sims.append(number_of_sims[-1]+1)
#    print(f'i {i} sim #{number_of_sims[-1]}')
    avg_earnings.append(earnings/number_of_sims[-1])
    
plt.plot(number_of_sims, avg_earnings,'.')
plt.xlabel('number of rolls')
plt.ylabel('average earnings')
```

### Craps

Now, we'll learn how to play craps. Craps is based off calculating the sum of two dice. If two dice with faces 1-6 are rolled, then the possible sums are 2-12. Here are the theoretical probabilities:

|  Sum |Probability   |
|---   |---   |
|2 |1/36  | 
|3 | 2/36 |
|4 |3/36  |
|5 | 4/36 |
|6 |5/36  |
|7 | 6/36 |
|8 | 5/36 |
|9 | 4/36 |
|10 | 3/36 |
|11 | 2/36 |
|12 | 1/36 |

Okay, so now you know the theoretical probabilities. Let's calculate the experimental probabilities.

### Craps: Simulate the dice rolls

Use the randint function twice to simulate two different dice rolls. Calculate the sum of the two dice. Create a list called sums that contains the sums from performing this simulation 10,000 times.

```python
sums = []
for i in range(10000):
    die1 = random.randint(1,6)
    die2 = random.randint(1,6)
    sums.append(die1+die2)
print(f"first ten rolls: {sums[0:10]}")
```

### Craps: Create a histogram of the dice rolls:

One of the most common types of bets in craps is a place bet. A place bet is betting on a number and wanting that number to come up before a 7 does. For example, if I bet on 4, then I am wanting a sum of 4 to come up before a sum of 7. I will continue to roll the dice until either a 4 or a 7 comes up, none of the other sums matter. What is the probability that you win this type of bet? 3/9, since there are 3 ways to win ((1,3), (2,2), or (3,1)) and 6 ways to lose ((1,6), (2,5), (3,4), (4,3), (5,2), (6,1)).

The theoretical probabilities of winning each place bet is below:

|  Sum |Probability   |
|---   |---   |
|4 |3/9  |
|5 | 4/10 |
|6 |5/11  |
|8 | 5/11 |
|9 | 4/10 |
|10 | 3/9 |

### Craps: Simulate winning percentage of Place 6 Bet
Now let's calculate the experimental probability of winning the Place 6 bet. Roll the dice 10,000 times and calculate the percentage of times that you win. Your decimal should be very close to the theoretical probability in the table above (5/11 = 0.4545)

Note: just because you rolled the dice 10,000 times does not mean that your number of wins + losses adds up to 10,000, because only a dice sum of 6 or 7 matters.



```python
wins = 0
mybet = 6
losses = 0
for i in range(10000):
        if sums[i] == 7:
            losses = losses + 1
        if sums[i] == mybet:
            wins = wins + 1

print('number of games', wins+losses)            
print('winning percentage', wins/(wins+losses))
```

Or a much more concise solution using counter:

```python
from collections import Counter

print(Counter(sums))

#count the number of 6's and divide by the sum of the 6's and 7's:
sixes = Counter(sums)[6]
total_games = Counter(sums)[6] + Counter(sums)[7]
print('winning percentage', sixes/total_games)
```
