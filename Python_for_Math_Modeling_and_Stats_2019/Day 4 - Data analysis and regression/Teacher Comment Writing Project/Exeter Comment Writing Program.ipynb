```python
import pandas as pd
df = pd.read_excel('exeter_student_data.xlsx')
df.head()
```

```python
with open('output.txt', 'w') as fout:
    for i in range(len(df.index)):
        
        string = ''
        #intro sentence about topics covered
        string = "In this quarter, we covered {} \n \n".format(df['Topics'][0])

        #student reflection - ask for reflection via Google Form and then copy and paste from Google forms summary column
        string = string + "{}, in your self reflection, you wrote: \n \n\'".format(df['First Name'][i])
        string = string + "{}\' \n \n".format(df['Student Reflection'][i])

        #overall performance
        if df['Letter Grade'][i] in ['A', 'A-', 'B+']:
            string = string + '{}, I am pleased with your progress this semester.'.format(df['First Name'][i]) 
        else:
            string = string + '{}, this quarter has been challenging for you but you are persevering.'.format(df['First Name'][i]) 

        string = string + ' Your homework average is {}%'.format(df['Homework'][i])

        #homework average
        if int(df['Homework'][i]) > 92:
            string = string + ' which indicates that your nightly dedication has been consistent.'
        else:
            string = string + ' which aligns with your self assessment that you should strive for more consistency in your nightly preparations.'

        #project, if any
        if 'Project Name' in df.columns:
            if int(df['Project Grade'][i])>89:
                string = string + ' In addition, your {} on {} was thorough and interesting.'.format(df['Project Name'][0], df['Project Comment'][i])
            else:
                string = string + ' In your {} on {}, you were missing a bit of the analysis.'.format(df['Project Name'][0], df['Project Comment'][i])    

        # habits of mind - larger discussion habits
        string = string + ' In the larger class discussion, '
        if df['Good Questions - see key'][i] == 0:
            string = string + 'you rarely asked questions, '
        if df['Good Questions - see key'][i] == 1:
            string = string + 'you were typically quiet, but I could see you following along intently, '
        if df['Good Questions - see key'][i] == 2:
            string = string + "you weren't afraid to ask questions when you needed extra clarification, "
        if df['Good Questions - see key'][i] == 3:
            string = string + 'the questions and comments that you posed were deep and insightful, '

        #habits of mind - peer collaboration
        string = string + 'and in small group work, '
        if df['Work with peers - see key'][i] == 0:
            string = string + 'you typically worked on your own.'
        if df['Work with peers - see key'][i] == 1:
            string = string + 'you worked collaboratively with your classmates to problem solve.'

        #habits of mind - extra help
        string = string + ' Outside of class, '
        if df['Stop by for extra help - see key'][i] == 0:
            string = string + 'you did not stop by for extra help very often, which would have been beneficial to your understanding of the material and would have decreased the stress that you felt towards the course I think.'
        if df['Stop by for extra help - see key'][i] == 1:
            string = string + 'you stopped by my office whenever you needed extra help.'
        if df['Stop by for extra help - see key'][i] == 2:
            string = string + "you have not needed to stop by for extra help, as you have understood the topics independently."


        #quiz average, if any
        if 'Quiz Average' in df.columns:
            string = string + '\n \nYour quiz average was {}%,'.format(df['Quiz Average'][i])
            if int(df['Quiz Average'][i]) > 88:
                string = string + ' which indicates that you usually pick up on new topics introduced quickly.'
            else:
                string = string + ' which indicates that sometimes you need more time to absorb new content before it really sticks.'


        #chapter tests
        #first figure out how many tests
        tests = [(int(df[column][i]), column) for column in df.columns if "Test" in column]
        string = string + ' You received a '
        for test in tests:
            string = string + '{}% on the {} and a '.format(test[0], test[1])
        if sum([test[0] for test in tests])/len(tests) > 84:
            string = string[:-7] + ', which aligns with your self-assessment that you were comfortable with most topics.'
        else:
            string = string[:-7] + ', which aligns with your self-assessment that you were not quite comfortable with all of the topics covered.'   

        #final exam, if any
        if 'Final Exam' in df.columns:
            string = string + ' Your final exam score was {}%, '.format(df['Final Exam'][i])
            if int(df['Final Exam'][i] > 85):
                string = string + "which indicates an ability to synthesize and retain most of the course content."
            else:
                string = string + "which indicates some trouble with synthesizing and retaining some of the course content."

        #overall grade
        string = string + '\n \nYour grade in the course is a {} for the quarter.'.format(df['Letter Grade'][i]) 
        if df['Letter Grade'][i] in ['A', 'A-', 'B+']:
            string = string + ' You should be proud of your growth as a mathematician in this course.'
        else:
            string = string + ' Please feel free to stop by for extra review sessions with me in order to help to solidify your understanding. I am happy to help!'

        #clean up language typos
        string = string.replace('a 8', 'an 8')
        string = string.replace('a A', 'an A')
        string = string + ' \n \n'

        print(string)
        fout.write('----------- \n')
        fout.write(string)
```
